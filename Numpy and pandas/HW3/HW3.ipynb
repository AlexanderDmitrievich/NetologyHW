```python
rating = pd.read_csv('C:/Users/User/Desktop/ФГУП НТЦ РХБГ 2021/Другое/Data Science/HW/Numpy Pandas/HW3/ml-latest-small/ratings.csv')
rating.head()
```

```python
joined = movie.merge(rating, on='movieId', how='left')
joined.head()
```

```python
def classify (data):
    if data['rating'] <= 2.0:
        return 'Низкий рейтинг'
    elif 2.0 < data['rating'] <= 4.0:
        return 'Средний рейтинг'
    else:
        return 'Высокий рейтинг'
```

```python
joined['class'] = joined.apply(classify, axis=1)
joined.head()
```

# Task 2

```python
kw = pd.read_csv('C:/Users/User/Desktop/ФГУП НТЦ РХБГ 2021/Другое/Data Science/HW/Numpy Pandas/HW3/ml-latest-small/keywords.csv')
kw.head()
```

```python
geo_data = {

'Центр': ['москва', 'тула', 'ярославль'],

'Северо-Запад': ['петербург', 'псков', 'мурманск'],

'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}
cities=[]
for values in geo_data.values():
    for i in values:
        cities.append(i)
cities
```

```python
def region(data):
    slovo = data['keyword'].split(' ')
    i = 0
    for word in slovo:
        if word in cities:
            for items in geo_data.items():
                if word in items[1]:
                    i+=1
                    return items[0]
    if i == 0:
        return 'undefined'
```

```python
kw['region'] = kw.apply(region, axis=1)
kw.head(10)
```

```python
kw[(kw['region']!='undefined')]
```

```python
years = []
i = 1950
while i <= 2010:
    years.append(i)
    i += 1 
```

```python
def production_year (data):
    for i in years:
        if str(i) in data['title']:
            return int(i)
    return ('1900')
```

```python
joined['year'] = joined.apply(production_year, axis=1)
```

```python
joined.head()
```

```python
res = joined.groupby('year').agg(['mean'])['rating'].sort_values('mean', ascending=False)
res
```
