# Task 2

```python
power = pd.read_csv('C:/Users/User/Desktop/ФГУП НТЦ РХБГ 2021/Другое/Data Science/HW/Numpy Pandas/HW2/power.csv')
power.head()
```

```python
power['country'].unique()
```

```python
power[ power['country'].str.contains('ia', case=False) ]['country'].unique()
```

```python
power[power.country.str.contains('|'.join(['Lat','Lit','Est']))].head(100)
```

```python
baltika = power[ (power['country']=='Lithuania') | (power['country']=='Latvia') | (power['country']=='Estonia') ]
baltika.head()
```

```python
year1 = baltika [baltika.year < 2010 ]
```

```python
year2 = year1 [ year1.year > 2005]
year2.head()
```

```python
quant = year2 [ year2.quantity > 0]
quant.head()
```

```python
cat = quant[ (quant['category']== 4 ) | ( quant['category']== 12 ) | (quant['category']== 21 ) ]
cat.head()
```

```python
summ = cat.groupby(['country']).sum()
summ.head()
```

```python
summ['quantity']
```
