Так как я сейчас продаю авто, попробуем сделать датасет с дальнейшим прогнозированием стоимости авто

```python
import numpy as np
import pandas as pd
import seaborn as sns

from sklearn.linear_model import LinearRegression
```

```python
n_samples = 10000

price = np.random.choice(range(500000,2000000), n_samples) #цена
age = np.random.choice(range(1,20), n_samples) # возраст
km = np.random.choice(range(0,100000), n_samples) # пробег
klass = np.random.choice((1,2,3), n_samples) # класс авто
owners = np.random.choice(range(1,5), n_samples) # количество собственников
country_list = ['Germany', 'Korea', 'China']
country = [np.random.choice(country_list) for sample in range(0,10000)] # страна происхождения
dtp = np.random.choice(range(1,5), n_samples) # количество ДТП
hp = np.random.choice(range(70,300), n_samples) # Лошадиные силы. В РФ с 250 л.с повышенный налог.
#Ограничение на 300 лс сделано специально, чтобы не было много авто с большим количеством лс
#Марку не писал специально, будем отталкиваться от страны происхождения и класса

data = pd.DataFrame({'price' : price,'age': age,'km': km,'klass': klass, 'owners': owners,'country': country,'dtp': dtp,'hp': hp})

data.loc[data["age"] < 5, "price"] += 2000000
data.loc[((data["age"] >= 5) & (data["age"] < 10 )), "price"] +=  1000000
data.loc[data["age"] > 15, "price"] -= 500000
                                                     
data.loc[((data["age"] >= 5) & (data["age"] <= 10 )), "km"] +=  75000
data.loc[((data["age"] > 10) & (data["age"] <= 15 )), "km"] +=  100000
data.loc[data["age"] > 15, "km"] += 150000
                                                               
data.loc[data["klass"] == 1, "price"] += 1000000
data.loc[data["klass"] == 2, "price"] += 500000

data.loc[data["owners"] == 1, "price"] += 500000 
data.loc[data["owners"] > 3, "price"] -= 300000

data.loc[data["country"] == "Germany", "price"] += 2000000 
data.loc[data["country"] == "China", "price"] -= 500000

data.loc[((data["km"] > 100000) & (data["km"] < 200000 )), "dtp"] += 1
data.loc[data["km"] > 200000, "dtp"] += 2

data.loc[data["dtp"] > 3, "price"] -= 300000

data.loc[data["hp"] > 250, "price"] -= 300000

data.loc[data["price"] < 250000, "price"] = 250000
```

```python
data.head(20)
```

```python
data.info()
```

Заменим категориальные данные на числовые

```python
data = pd.get_dummies(data, columns = ['country'])
data.head()
```

Попробуем внести NaN и выбросы

```python
# NaN
for i in range(100):
    data.loc[np.random.randint(0,1000), np.random.choice(data.columns)] = np.nan

# Выбросы
for i in range(10):
    data.loc[np.random.randint(0,1000), 'age'] = 100

for i in range(10):
    data.loc[np.random.randint(0,1000), 'km'] = 1000000

for i in range(10):
    data.loc[np.random.randint(0,1000), 'hp'] = 1000
```

```python
data.head(20)
```

```python
data.info()
```

Заменим NaN на средние значения

```python
for i in data:
    print (round(data[i].mean()))
```

```python
data = data.fillna(round(data.mean()))
data.head(20)
```

```python
data.info()
```

```python
X = data[data.columns[1:]]
X.head()
```

```python
y = data['price']
y.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model = LinearRegression()

model.fit(X_train, y_train)

model.score(X_test, y_test)
```
