Задание

При фиксированном seed=42 поэкспериментируйте с параметрами алгоритма differential_evolution: strategy, popsize, mutation.

Постройте графики количества итераций (nit) оптимизации функции ackley от значения параметра.

Подробнее о результате выдачи тут.

```python
import numpy as np
from scipy.optimize import differential_evolution
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def ackley(x):
    arg1 = -0.2 * np.sqrt(0.5 * (x[0] ** 2 + x[1] ** 2))
    arg2 = 0.5 * (np.cos(2. * np.pi * x[0]) + np.cos(2. * np.pi * x[1]))
    return -20. * np.exp(arg1) - np.exp(arg2) + 20. + np.e

bounds = [(-10, 10), (-10, 10)]

result = differential_evolution(ackley, bounds, seed=42)
result
```

```python
list_strategy = ['best1bin', 
                 'best1exp', 
                 'rand1exp', 
                 'randtobest1exp',
                 'currenttobest1exp', 
                 'best2exp', 
                 'rand2exp', 
                 'randtobest1bin',
                 'currenttobest1bin',
                 'best2bin',
                 'rand2bin',
                 'rand1bin']

list_nit = []

for strategy in list_strategy:
    result = differential_evolution(ackley, bounds, seed=42, strategy=strategy)
    list_nit.append(result.nit)

plt.figure(figsize=(20, 10))    
plt.bar(list_strategy, list_nit)
plt.show()
```

```python
list_nit = []
list_popsize = []

for size in range(10, 100, 5):
    result = differential_evolution(ackley, bounds, seed=42, popsize=size)
    list_nit.append(result.nit)
    list_popsize.append(size)

plt.figure(figsize=(20, 10))    
plt.plot(list_popsize, list_nit)
plt.show()
```

```python
list_nit = []
list_mutation = []

for mutation in np.arange(0.1, 2, 0.01):
    result = differential_evolution(ackley, bounds, seed=42, mutation=mutation)
    list_nit.append(result.nit)
    list_mutation.append(mutation)
    
plt.figure(figsize=(20, 10))    
plt.plot(list_mutation, list_nit)
plt.show()
```
