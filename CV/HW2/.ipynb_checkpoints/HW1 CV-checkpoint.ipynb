```python
%matplotlib inline
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Загружаем исходные данные

```python
train = np.loadtxt('train.csv', delimiter=',', skiprows=1)
test = np.loadtxt('test.csv', delimiter=',', skiprows=1)
```

```python
# сохраняем разметку в отдельную переменную
train_label = train[:, 0]
# приводим размерность к удобному для обаботки виду
train_img = np.resize(train[:, 1:], (train.shape[0], 28, 28))
test_img = np.resize(test, (test.shape[0], 28, 28))
```

```python
train_img.shape
```

## Визуализируем исходные данные

```python
fig = plt.figure(figsize=(20, 10))
for i, img in enumerate(train_img[0:5], 1):
    subplot = fig.add_subplot(1, 7, i)
    plt.imshow(img, cmap='gray');
    subplot.set_title('%s' % train_label[i - 1]);
```

## Вычисляем X и Y составляющие градиента с помощью оператора Собеля

```python
train_sobel_x = np.zeros_like(train_img)
train_sobel_y = np.zeros_like(train_img)
for i in range(len(train_img)):
    train_sobel_x[i] = cv2.Sobel(train_img[i], cv2.CV_64F, dx=1, dy=0, ksize=3)
    train_sobel_y[i] = cv2.Sobel(train_img[i], cv2.CV_64F, dx=0, dy=1, ksize=3)
```

```python
test_sobel_x = np.zeros_like(test_img)
test_sobel_y = np.zeros_like(test_img)
for i in range(len(test_img)):
    test_sobel_x[i] = cv2.Sobel(test_img[i], cv2.CV_64F, dx=1, dy=0, ksize=3)
    test_sobel_y[i] = cv2.Sobel(test_img[i], cv2.CV_64F, dx=0, dy=1, ksize=3)
```

## Вычисляем угол и длину вектора градиента

```python
train_g, train_theta = cv2.cartToPolar(train_sobel_x, train_sobel_y)
```

```python
test_g, test_theta = cv2.cartToPolar(test_sobel_x, test_sobel_y)
```

```python
fig = plt.figure(figsize=(20, 10))
for i, img in enumerate(train_g[:5], 1):
    subplot = fig.add_subplot(1, 7, i)
    plt.imshow(img, cmap='gray');
    subplot.set_title('%s' % train_label[i - 1]);
    subplot = fig.add_subplot(3, 7, i)
    plt.hist(train_theta[i - 1].flatten(),
             bins=16, weights=train_g[i - 1].flatten())
```

## Вычисляем гистограммы градиентов

```python
# Гистограммы вычисляются с учетом длины вектора градиента
train_hist = np.zeros((len(train_img), 16))
for i in range(len(train_img)):
    hist, borders = np.histogram(train_theta[i],
                                 bins=16,
                                 range=(0., 2. * np.pi),
                                 weights=train_g[i])
    train_hist[i] = hist
```

```python
test_hist = np.zeros((len(test_img), 16))
for i in range(len(test_img)):
    hist, borders = np.histogram(test_theta[i],
                                 bins=16,
                                 range=(0., 2. * np.pi),
                                 weights=test_g[i])
    test_hist[i] = hist
```

## Нормируем вектор гистограммы

```python
# По умолчанию используется L2 норма
train_hist = train_hist / np.linalg.norm(train_hist, axis=1)[:, None]
```

```python
test_hist = test_hist / np.linalg.norm(test_hist, axis=1)[:, None]
```

## Разбиваем выборку на обучение и валидацию

```python
from sklearn.model_selection import train_test_split
y_train, y_val, x_train, x_val = train_test_split(
    train_label, train_hist, test_size=0.2,
    random_state=42)
```

## Используем свою модель

```python
import warnings

warnings.filterwarnings('ignore')
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import ExtraTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.neighbors import RadiusNeighborsRegressor
```

```python
lr_model = LinearRegression()
dt_model = DecisionTreeRegressor()
et_model = ExtraTreeRegressor()
kn_model = KNeighborsRegressor()
rn_model = RadiusNeighborsRegressor()
```

```python
lr_params = {'fit_intercept':[True, False], 'normalize':[True, False]}
dt_params = {'min_samples_split':[x for x in range(1,10)], 'min_samples_leaf':[x for x in range(1,10)]}
et_params = {'min_samples_split':[x for x in range(1,10)], 'min_samples_leaf':[x for x in range(1,10)]}
kn_params = {'n_neighbors': [x for x in range(1,10)], 'weights': ['uniform', 'distance'], 'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute']}
rn_params = {'radius': [x for x in np.arange(0.0, 5.1, 0.2)], 'weights': ['uniform', 'distance'], 'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute']}
```

```python
list_model = [lr_model, dt_model, et_model, kn_model, rn_model]
list_param = [lr_params, dt_params, et_params, kn_params, rn_params]
```

```python
from sklearn.model_selection import GridSearchCV

for model, grid_param in zip(list_model, list_param):
    grid = GridSearchCV(model, grid_param, cv=3, scoring='r2')
    grid.fit(x_train, y_train)
    print(model)
    print("Best score ->", grid.best_score_)
    print("Best params ->", grid.best_params_)
    print("Best estimator ->", grid.best_estimator_)
    for k in grid.cv_results_:
        print(k, ":", grid.cv_results_[k][0])
    
    print("---------------------------------------------------------------")
```

Лучший результат у KNeighborsRegressor(), но этого все равно мало. Пробуем дальше.

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier()
```

```python
rf_model.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
print('Accuracy: %s' % accuracy_score(y_val, rf_model.predict(x_val)))
```

RandomForestClassifier() показал хороший результат. Оставляем его.

## Визаулизируем на тестовой выборке

```python
result = rf_model.predict(test_hist)
```

```python
fig = plt.figure(figsize=(20, 10))
for i, img in enumerate(test_img[0:7], 1):
    subplot = fig.add_subplot(1, 7, i)
    plt.imshow(img, cmap='gray');
    subplot.set_title('%s' % result[i - 1]);
```

```python
with open('submit.csv', 'w') as dst:
    dst.write('ImageId,Label\n')
    for i, p in enumerate(result, 1):
        dst.write('%s,%s\n' % (i, str(p).replace('.0', '')))
```
