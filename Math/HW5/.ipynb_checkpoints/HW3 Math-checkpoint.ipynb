Дан объект в  2D пространстве

```python
import numpy as np
import matplotlib.pyplot as plt

A = [
[-100, -100, 1],
[-100, 100, 1],
[100, 100, 1],
[0, 0, 1],
[100, -100, 1],
[-100, -100, 1]
]

A = np.array(A)

x = A[:,0]
y = A[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200]) 
```

При помощи линейных отображений:

Уменьшить объект в два раза

Повернуть на 130 градусов

Отразить объект относительно прямой y=x (поможет матрица [[0, 1, 0], [1, 0, 0], [0, 0, 1]])

```python
def draw_object(AA):
    x = AA[:,0]
    y = AA[:,1]
    plt.plot(x, y)
    plt.ylim([-500, 500])
    plt.xlim([-500, 500])
    plt.grid()
```

```python
A2 = A/2
draw_object(A2)
```

```python
from random import randint
from scipy import linalg
import math
from math import cos, sin
theta = 130*math.pi/180
```

```python
rotation_matrix = np.array([[np.cos(theta), -np.sin(theta)],[np.sin(theta),  np.cos(theta)]])
rotation_matrix
```

```python
matr = np.array([x,y])
```

```python
rot = np.dot(rotation_matrix, matr)
```

```python
plt.plot(rot[0], rot[1])
plt.ylim([-200, 200])
plt.xlim([-200, 200])
plt.grid()
```

```python
refl = np.array([[0, 1, 0], [1, 0, 0], [0, 0, 1]])
```

```python
refl_res = np.dot(A,refl)
```

```python
draw_object(refl_res)
```

Найдите спектральное разложение матрицы:

m = np.array([[1, 2],
              [2, 3]])

```python
m = np.array([[1, 2], [2, 3]]) 
print(m)
```

```python
w, v = np.linalg.eig(m) # Находим собственные значения и вектора
v_T = np.linalg.inv(v) # получается обратную матрицу матрице собственных векторов
L = np.diag(w) # Диагональная матрица с собственными значениями
print(v)
print(L)
print(v_T)

print('Проверка:\n', v @ L @ v_T)
```
