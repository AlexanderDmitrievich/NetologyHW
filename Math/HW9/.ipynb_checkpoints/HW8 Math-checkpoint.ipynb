Задание 1
В магазин привезли устройства с 3-х разных предприятий.

Соотношение устройств следующее: 20% - продукция первого предприятия, 30% - продукция второго предприятия, 50% - продукция третьего предприятия; далее, 10% продукции первого предприятия высшего сорта, на втором предприятии - 5% и на третьем - 20% продукции высшего сорта.

Найти вероятность того, что случайно купленная новая продукция окажется высшего сорта.

```python
pA1 = 0.2 # Вероятность купить продукцию 1 преприятия
pB1_A1 = 0.1 # Вероятность купить продукцию высшего сорта при условии что купили продукцию 1 предприятия

pA2 = 0.3 # Вероятность купить продукцию 2 преприятия
pB2_A2 = 0.05 # Вероятность купить продукцию высшего сорта при условии что купили продукцию 2 предприятия

pA3 = 0.5 # Вероятность купить продукцию 3 преприятия
pB3_A3 = 0.2 # Вероятность купить продукцию высшего сорта при условии что купили продукцию 3 предприятия

# вероятность того, что случайно купленная новая продукция окажется высшего сорта
P = pB1_A1*pA1+pB2_A2*pA2+pB3_A3*pA3
print("Вероятность купить продукцию высшего сорта ->", P)
```

Задание 2

Пусть брошены 3 уравновешенные монеты.

Рассмотрим 3 события:

A1 - монеты 1 и 2 упали одной и той же стороной;

A2 - монеты 2 и 3 упали одной и той же стороной;

A3 - монеты 1 и 3 упали одной и той же стороной.

Покажите, почему эти 3 события (A1, A2, A3) являются попарно независимыми, но не являются независимыми в совокупности.

```python
pA1 = 0.5 # Вероятность наступления события А1
pA2 = 0.5 # Вероятность наступления события А2
pA3 = 0.5 # Вероятность наступления события А2

pA1A2 = 0.5*0.5 # Вероятность наступления события А1 и А2 одновременно 
pA1A3 = 0.5*0.5 # Вероятность наступления события А1 и А3 одновременно
pA2A3 = 0.5*0.5 # Вероятность наступления события А2 и А3 одновременно
# В обратные стороны будет тоже самое - поэтому не стал отдельно считать

# Дальне считал условные вероятности

pA1_A2 = pA1A2/pA2 # Вероятность А1 при наступлении А2
pA1_A3 = pA1A3/pA3 # Вероятность А1 при наступлении А3

pA2_A1 = pA1A2/pA1 # Вероятность А2 при наступлении А1
pA2_A3 = pA2A3/pA3 # Вероятность А2 при наступлении А3

pA3_A1 = pA1A3/pA1 # Вероятность А3 при наступлении А1
pA3_A2 = pA2A3/pA2 # Вероятность А3 при наступлении А2

# Проверяем: если вероятность наступления события равно вероятности наступления этого события при условии наступления другого
# То события независимые
# Проверяем все пары
if pA1 == pA1_A2 and pA1 == pA1_A3 and pA2 == pA2_A1 and pA2 == pA2_A3 and pA3 == pA3_A1 and pA3 == pA3_A2:
    print("События попарно независимы")

pA1A2A3 = 0.25 # Вероятность наступления всех трех событий сразу

# если вероятность наступления всех трех событий равно произвдению вероятностей каждого события
# то события являются независимыми в совокупности
if pA1A2A3 != pA1*pA2*pA3:
    print("События не являются независимыми в совокупности")
    
```

Задание 3

Изучите 3 распределения из библиотеки scipy: Нормальное, Экспоненциальное, Стьюдента. Используя функцию плотности распредления (pdf), постройте пять графиков плотностей для каждого распределения при разных параметрах. Запишите ваши наблюдения: как и на что влияют параметры у каждого распределения.

Дополнительная информация тут.

```python
import numpy as np
import scipy.stats as sts
import matplotlib.pyplot as plt
```

```python
# Нормальное распределение

mu_list = range(1,6)
sigma_list = np.arange(0.1, 0.6, 0.1)

for mu, sigma in zip(mu_list, sigma_list):
    norm_rv = sts.norm(loc=mu, scale=sigma)
    x = np.linspace(0,6,100)
    pdf = norm_rv.pdf(x)
    plt.plot(x, pdf)

plt.show()
```

```python
# Экспоненциальное распределение

for lam in np.arange(0.1, 0.6, 0.1):
    exp_rv = sts.expon(lam)
    x = np.linspace(0,1,100)
    pdf = exp_rv.pdf(x)
    plt.plot(x, pdf)

plt.show()
```

```python
# Т распределение Стьюдента

for n in range(10, 1000, 10):
    t_rv = sts.t(n)
    x = np.linspace(-5,5,1000)
    pdf = t_rv.pdf(x)
    plt.plot(x, pdf)

plt.show()
```

Нормальное распределение. При изменении мю среднее значение смещается в сторону изменения мю. При изменении средне квадратичного отклонения разброс от среднего также изменяется.

Экспоненциальное распределение. При изменении параметра лямбда передвигается вершина графика в сторону его увеличения. Соответственно среднее значение случайной величины увеличивается.

Т распределение Стьюдента. Увеличение выборки случайных величин практически не изменяет график. Т.е. среднее значение остается примерно таким же и отклонение от него почти не меняется.
