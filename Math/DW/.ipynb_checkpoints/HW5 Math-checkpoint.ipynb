```python
import numpy as np
from sympy import *
```

```python
x, y = symbols('x y')
fxy = 2*x**2*y**3+1/x+y**2*x+7

dx_fxy = diff(fxy, x)
print("Частная производная по x: {}".format(dx_fxy))

dy_fxy = diff(fxy, y)
print("Частная производная по y: {}".format(dy_fxy))
```

```python
x, y = symbols('x y')
fxy = x**2*y-sin(x*y)+cos(x**2)+6*y

dx_fxy = diff(fxy, x)
print("Частная производная по x: {}".format(dx_fxy))

dy_fxy = diff(fxy, y)
print("Частная производная по y: {}".format(dy_fxy))
```

в точке (1,2)

```python
from scipy.optimize import approx_fprime

def fx(x, c0, c1):
    return c0*x[0]**2*x[1]**3+1/x[0]+x[1]**2*x[0]+c1

x = np.array([1, 2])
c0, c1 = (2, 7)
eps=np.sqrt(np.finfo(float).eps)

print("Значение градиента в точке (1, 2) ->", approx_fprime(x, fx, [eps, eps], c0, c1))
```

```python
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
import numpy as np
ax = axes3d.Axes3D(plt.figure())
i = np.arange(-10, 10, 0.01)
x1, x2 = np.meshgrid(i, i)
Z = x2**2 - ((x1**2)*x2) + (2*x1*x2)
#ax = plt.axes(projection='3d')
#ax.plot_wireframe(x1, x2, Z, color ='red', rstride=10, cstride=10) 
ax.plot_surface(x1, x2, Z, rstride=10, cstride=10,
                cmap='viridis')
ax.set_title('Трехмерный график функции') 
plt.show()
```
