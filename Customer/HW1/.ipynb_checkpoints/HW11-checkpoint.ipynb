```python
import pandas as pd
import numpy as np
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns

%config InlineBackend.figure_format = 'retina'
from pylab import rcParams
rcParams['figure.figsize'] = (12, 9)
```

```python
data = pd.read_excel('C:/Users/User/Desktop/ФГУП НТЦ РХБГ 2022/Другое/Data Science 2022/HW/Python/ML/HW11/Homework/geo.xlsx')
data.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_scaled = scaler.fit_transform(data[['x','y']])
```

```python
from sklearn.cluster import DBSCAN, KMeans
```

```python
kmn = KMeans()
clusters = kmn.fit_predict(X_scaled)
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=clusters, s=60)
```

```python
k_inertia = []
ks = range(1,11)

for k in ks:
    clf_kmeans = KMeans(n_clusters=k)
    clusters_kmeans = clf_kmeans.fit_predict(X_scaled, )
    k_inertia.append(clf_kmeans.inertia_)
    
plt.plot(ks, k_inertia)
```

```python
diff = np.diff(k_inertia)

plt.plot(ks[1:], diff)
```

```python
diff_r = diff[1:] / diff[:-1]
plt.plot(ks[1:-1], diff_r)
```

```python
k_opt = ks[np.argmin(diff_r)+1]
print(k_opt)
```

```python
kmeans_clust = KMeans(n_clusters=3)
clusters = kmeans_clust.fit_predict(X_scaled)
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=clusters, s=60)
```

```python
comment_data = pd.read_excel('C:/Users/User/Desktop/ФГУП НТЦ РХБГ 2022/Другое/Data Science 2022/HW/Python/ML/HW11/Homework/geo_comment.xlsx')
comment_data.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(comment_data['comment'])
```

```python
k_inertia = []
ks = range(1,11)
i = 1

for k in ks:
    clf_kmeans = KMeans(n_clusters=k)
    clusters_kmeans = clf_kmeans.fit_predict(X, )
    k_inertia.append(clf_kmeans.inertia_)
    i += 1
    
plt.plot(ks, k_inertia)
```

```python
diff = np.diff(k_inertia)

plt.plot(ks[1:], diff)
```

```python
diff_r = diff[1:] / diff[:-1]
plt.plot(ks[1:-1], diff_r)
```

```python
k_opt = ks[np.argmin(diff_r)+1]
k_opt
```

```python
clf_kmeans = KMeans(n_clusters=8)
clusters_kmeans = clf_kmeans.fit_predict(X, )
```

```python
comment_data['term'] = clusters_kmeans
```
