Самый коррелирующий признак - Wilderness_Area_3. Попробуем построить модель только с ним и сравнить результаты.

```python
feature_names = ['Elevation', 'Aspect', 'Slope', 'Horizontal_Distance_To_Hydrology', 'Vertical_Distance_To_Hydrology', 
'Horizontal_Distance_To_Roadways', 'Hillshade_9am', 'Hillshade_Noon', 'Hillshade_3pm', 'Horizontal_Distance_To_Fire_Points', 
'Wilderness_Area_0', 'Wilderness_Area_1', 'Wilderness_Area_2', 'Wilderness_Area_3', 'Soil_Type_0', 'Soil_Type_1', 
'Soil_Type_2', 'Soil_Type_3', 'Soil_Type_4', 'Soil_Type_5', 'Soil_Type_6', 'Soil_Type_7', 'Soil_Type_8', 'Soil_Type_9', 
'Soil_Type_10', 'Soil_Type_11', 'Soil_Type_12', 'Soil_Type_13', 'Soil_Type_14', 'Soil_Type_15', 'Soil_Type_16', 
'Soil_Type_17', 'Soil_Type_18', 'Soil_Type_19', 'Soil_Type_20', 'Soil_Type_21', 'Soil_Type_22', 'Soil_Type_23', 
'Soil_Type_24', 'Soil_Type_25', 'Soil_Type_26', 'Soil_Type_27', 'Soil_Type_28', 'Soil_Type_29', 'Soil_Type_30', 
'Soil_Type_31', 'Soil_Type_32', 'Soil_Type_33', 'Soil_Type_34', 'Soil_Type_35', 'Soil_Type_36', 'Soil_Type_37', 
'Soil_Type_38', 'Soil_Type_39']
```

```python
feature_idx = {}
i = 0
for name in feature_names:
    feature_idx[name] = i
    i += 1
```

```python
X_t_1 = X_train[:,feature_idx['Wilderness_Area_3']]
X_t_1 = X_t_1.reshape(-1, 1)
X_test_1 = X_test[:,feature_idx['Wilderness_Area_3']]
X_test_1 = X_test_1.reshape(-1, 1)

clf1 = DecisionTreeClassifier(max_depth=10)
clf1.fit(X_t_1, y_train)
clf1.score(X_test_1,y_test)
```

Качество ниже, чем было при логистической регрессии. Попробуем добавлять признаки и смотреть на результат.

```python
def number_to_score(feature_order):
    scores = {}
    feature_list = []
    k = 0
    for i in range(0,(len(feature_order))):
        feature_list = []
        if i > -1:
            for k in range(0,i+1):
                if feature_order[k] != 'Cover_Type':
                    feature_list.append(feature_idx[feature_order[k]])

            if feature_list != []:
                X_train_1 = X_train[:,feature_list]
                X_test_1 = X_test[:,feature_list]

                clf1 = DecisionTreeClassifier(max_depth=7)
                clf1.fit(X_train_1, y_train)
                scores[i+1] = clf1.score(X_test_1,y_test)
    return scores
```

```python
feature_ordered = {}
for i in range(0,len(sorted_corr)):
    feature_ordered[i] = sorted_corr.index[i]
```

```python
scores_list = number_to_score(feature_ordered)
```

```python
plt.figure(figsize=(15,10))
plt.plot(scores_list.keys(),scores_list.values())
plt.xlabel('Количество признаков')
plt.ylabel('Качество')
```

При увеличении количества признаков улучшается качество. Попробуем RandomForest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier()
model.fit(X_train, y_train)
```

```python
feature_importances = {}
for i in range(len(feature_names)):
    feature_importances[feature_names[i]] = model.feature_importances_[i]
```

```python
feature_importances = dict(sorted(feature_importances.items(), key=lambda item: item[1],reverse = True))
feature_importances
```

```python
feature_ordered = {}
i = 0

for key in list(feature_importances.keys()):
    feature_ordered[i] = key
    i += 1
    
scores_list = number_to_score(feature_ordered)
```

```python
plt.figure(figsize=(15,10))
plt.plot(list(scores_list.keys())[:],list(scores_list.values())[:])
plt.xlabel('Количество признаков')
plt.ylabel('Качество')
```

В данном случае примерно после 24 признака улучшения модели не было. Теперь попробуем жадный отбор по максимальному нарастанию.

```python
feature_idx_reducing = feature_idx.copy()
score_to_num = {}
feature_list_full = []
max_score = 0
```

```python
for i in range(0,len(feature_idx)):
    
    feature_chosen = ''
    
    for key,value in feature_idx_reducing.items():
    
        feature_list = feature_list_full.copy()
        feature_list.append(value)

        X_train_1 = X_train[:,feature_list]
        X_test_1 = X_test[:,feature_list]

        if len(feature_list) < 2:
            X_train_1 = X_train_1.reshape(-1, 1)
            X_test_1 = X_test_1.reshape(-1, 1)

        clf = DecisionTreeClassifier(max_depth=7)
        clf.fit(X_train_1, y_train)
        score = clf.score(X_test_1,y_test)
        
        if score > max_score:
            max_score = score
            feature_chosen = value
            feature_chosen_key = key
            score_to_num[len(feature_list)] = max_score
    
    if feature_chosen == '':
        print(f'Max score: {max_score}. Number of features: {len(feature_list)}')
        break
    else:
        feature_list_full.append(feature_chosen)
        feature_idx_reducing.pop(feature_chosen_key)
```

```python
plt.figure(figsize=(15,10))
plt.plot(list(score_to_num.keys())[:],list(score_to_num.values())[:])
plt.xlabel('Количество признаков')
plt.ylabel('Качество')
```

В данном случае примерно после 18 признака улучшения модели не было. Теперь попробуем жадный отбор по минимальному падению.

```python
feature_idx_reducing = feature_idx.copy()
score_to_num = {}
feature_list_full = list(feature_idx.values())
prev_score = 0
delta_score = 1e-3
score = 0
max_score = 0
```

```python
for i in range(0,len(feature_idx)):

    feature_chosen = ''
    
    prev_score = max_score
    max_score = 0
    
    for key,value in feature_idx_reducing.items():
        

        feature_list = feature_list_full.copy()
        feature_list.remove(value)

        X_train_1 = X_train[:,feature_list]
        X_test_1 = X_test[:,feature_list]

        if len(feature_list) < 2:
            X_train_1 = X_train_1.reshape(-1, 1)
            X_test_1 = X_test_1.reshape(-1, 1)

        clf = DecisionTreeClassifier(max_depth=7)
        clf.fit(X_train_1, y_train)
        score = clf.score(X_test_1,y_test)
        
        if score > max_score:
            
            feature_chosen = value
            feature_chosen_key = key
            max_score = score
            score_to_num[len(feature_list)] = max_score

    
    if prev_score - max_score > delta_score:
        print(f' Number of features: {len(feature_list)}')
        break
    else:
        if feature_chosen != '':
            feature_list_full.remove(feature_chosen)
            feature_idx_reducing.pop(feature_chosen_key)
```

```python
plt.figure(figsize=(15,10))
keys = list(score_to_num.keys())
values = list(score_to_num.values())
plt.plot(keys,values)
plt.xlabel('Количество признаков')
plt.ylabel('Качество')
```
