The dataset was downloaded from the UCI Machine Learning Repository.

The two datasets are related to red and white variants of the Portuguese "Vinho Verde" wine. The reference [Cortez et al., 2009]. Due to privacy and logistic issues, only physicochemical (inputs) and sensory (the output) variables are available (e.g. there is no data about grape types, wine brand, wine selling price, etc.).

These datasets can be viewed as classification or regression tasks. The classes are ordered and not balanced (e.g. there are munch more normal wines than excellent or poor ones). Outlier detection algorithms could be used to detect the few excellent or poor wines. Also, we are not sure if all input variables are relevant. So it could be interesting to test feature selection methods.

Two datasets were combined and few values were randomly removed.

## 2.5 Взаимосвязи между переменными

```python
def plot_correlation_map( data ):
    corr = data.corr()
    _ , ax = plt.subplots( figsize =( 12 , 10 ) )
    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )
    _ = sns.heatmap(
        corr, 
        cmap = cmap,
        square=True, 
        cbar_kws={ 'shrink' : .9 }, 
        ax=ax, 
        annot = True, 
        annot_kws = { 'fontsize' : 12 }
    )
    
plot_correlation_map(data)
```

У колонки quality в принципе нет сильной корреляции ни с одним из признаков, поэтому будем использовать все.

## 3. Data preparation

В главе 2 мы увидели, что у нас есть незаполненные значения и выбросы. Уберем их:
#### 3.1 Nan:

```python
data1 = data.dropna()
data1.info()
data1.to_csv('data1.csv')
```

#### 3.2 Выбросы:

```python
data1.loc[data1['residual sugar'] > 20, 'residual sugar'] = data1['residual sugar'].mean()
data1.loc[data1['free sulfur dioxide'] > 100, 'free sulfur dioxide'] = data1['free sulfur dioxide'].mean()
data1.loc[data1['total sulfur dioxide'] > 300, 'total sulfur dioxide'] = data1['total sulfur dioxide'].mean()
```

```python
data1.describe()
```

```python
data2 = data1
data2.to_csv('data2.csv')
```

### 3.3 OneHotEncoding

### 3.4 Создадим датасет

```python
from sklearn.model_selection import train_test_split

train = data3[0:4000]
test = data3[4001:]

X = train.drop(columns=['quality'])
y = train['quality']

X_test = test.drop(columns=['quality'])

X_train , X_valid , y_train , y_valid = train_test_split( X , y , test_size = 0.3, random_state=42)
```

### 3.5 Важность признаков

```python
def plot_model_var_imp( model , X , y ):
    imp = pd.DataFrame( 
        model.feature_importances_  , 
        columns = [ 'Importance' ] , 
        index = X.columns 
    )
    imp = imp.sort_values( [ 'Importance' ] , ascending = True )
    imp[ : 10 ].plot( kind = 'barh' )
    print (model.score( X , y ))
    
def plot_variable_importance( X , y ):
    tree = DecisionTreeClassifier( random_state = 42 )
    tree.fit( X , y )
    plot_model_var_imp( tree , X , y )
```
