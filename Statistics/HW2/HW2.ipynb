```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sbn
```

Для анализа использовался файл со списком лучших фильмов, номинированных или получивших Оскар

```python
names = ['Film', 'Studio', 'Award', 'Year', 'Movie Time', 'Genre', 'IMDB Rating', 'IMDB Votes'] 
csv = 'C:/Users/User/Desktop/ФГУП НТЦ РХБГ 2021/Другое/Data Science/HW/Python/Statistics/HW2/oscars_df_kaggle.csv'
movies = pd.read_csv(csv, header = 0, usecols = [1, 3, 4, 5, 6, 7, 8, 9], names = names)
movies
```

```python
movies.info()
```

IMDB Votes типа object, хотя должен быть типа Int. 

```python
# Избавляемся от запятых, чтобы можно было переделать тип

for i in movies['IMDB Votes']:
    if ',' in i:
        f = i.replace(",","")
        movies['IMDB Votes'] = movies['IMDB Votes'].replace(i, f)
        print (movies['IMDB Votes'])
```

```python
movies['IMDB Votes'] = movies['IMDB Votes'].astype('Int64')
```

```python
movies.info()
```

```python
movies
```

Построим график среднего рейтинга IMDB по годам

```python
imdb_year = movies[['IMDB Rating', 'Year']].groupby('Year').mean()
imdb_year
```

```python
plt.plot(imdb_year.index, imdb_year['IMDB Rating'])
plt.title('Рейтинг IMDB по годам')
plt.xlabel('Год')
plt.ylabel('Рейтинг')
```

На этом графике мы видим сильный провал в течение первых лет наблюдений, затем достаточно быстрый подъем рейтингов. С 1940 по 1980 гг. видна слабая положительная динамика общего тренда. После 1980 г. средний рейтинг составляет примерно 7.7 с отклонением от среднего сзначения примерно на 0,7 п.

Построим график зависимости рейтинга от продолжительности фильма

```python
rating_time = movies[['IMDB Rating', 'Movie Time']].groupby('IMDB Rating').mean()
rating_time
```

```python
rating_time.plot(kind='bar', title='3ависимость рейтинга от продолжительности фильма',
                             xlabel='Рейтинг',
                             ylabel='Продолжительность, мин',
                             color = 'green')
```

На данном графике прослеживается увеличение среднего рейтинга с увеличением продолжительности фильма. Соответственно, чем дольше идет фильм, тем выше его рейтинг. Фильмы с продолжительностью от 150 минут имеют оценку близкую к 9, а фильмы до 100 минут имеют оценки ниже 6.5.

Посмотрим сколько голосов отдано за фильмы-номинанты, а сколько за победителей

```python
aw = movies['Award'].value_counts()
vo = movies['IMDB Votes'].groupby(movies['Award']).sum()
award_vote = vo / aw
award_vote
```

```python
award_vote.plot(kind='pie')
plt.legend()
```

Мы посчитали сколько в среднем отдано голосов за фильмы-победители и за фильмы-номинанты. За фильмы-победители голосовали практически вдвое больше, чем за фильмы-номинанты. 

# Для построения более сложных визуализаций данного датафрейма будет недостаточно, поэтому следующие визуализации будут строиться на основе рейтинга World Happiness Report 2021

```python
csv = 'C:/Users/User/Desktop/ФГУП НТЦ РХБГ 2021/Другое/Data Science/HW/Python/Statistics/HW2/world-happiness-report-2021.csv'
df = pd.read_csv(csv, header = 0)
df
```

```python
df[['Social support', 'Freedom to make life choices', 'Perceptions of corruption']].plot(kind='box')
```

Видно, что все три показателя являются действительно важными показателями для населения. На 1 месте находится соц. поддержка, на 2-ом с небольшим отставанием свободна выбора. Коррупция, очевидно, не так важна как предыдущие показатели, к тому же сопровождается большим количеством выбросов, что говорит о том, что в разных странах разное отношения к данному явлению.

```python
top_country = df.head(50)

heatmap=plt.figure(figsize=(16, 20)) 

sbn.heatmap(top_country.set_index('Country name')[['Social support', 'Freedom to make life choices', 'Perceptions of corruption']])
```

Данная визуализация отображает цевтом распределение значений от меньшего к большему, тоесть, чем выше показатель, тем светлее ячейка. Однако, так как уровень коррупции определяется наоборот (чем ниже тем лучше) нам стоит смотреть на то насколько светлые первые 2 ячейки и насколько темная третья ячейка, чтобы понять уровень рассматриваемых показателей в стране.

```python
sbn.pairplot(df[['Social support', 'Freedom to make life choices', 'Perceptions of corruption']])
```

Данная визуализация показывает распределение значений рассматриваемых показателей. Кучность точек и гистограммы отображают какие значения распространены больше всего. В социальной поддержке большинство рассматриваемых стран имеют значение показателя выше 0,7, в свободе выбора выше 0,5, а в коррупции выше 0,6 (что не очень хорошо, так как самым низким уровнем коррупции могут похвастать лишь небольшое количество стран)
